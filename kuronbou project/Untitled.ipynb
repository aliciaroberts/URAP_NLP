```python
# libraries

import numpy as np
import pandas as pd
import re
```

```python
negro_df = pd.read_json('../../data/output_negro.json')

negro_df
```

```python
sora_df = pd.read_csv('../../data/sora_sources.csv').dropna() # empty rows not needed

sora_df.head()
```

```python
sora_negro_df = sora_df[sora_df['Searched word'].str.contains('Negro')] # accounts for white spaces

sora_negro_df.head()
```

```python
# now let's merge these tables by the link:


negro_complete_df = sora_negro_df.merge(right = negro_df, # merge with Ryan's table
                                        left_on= 'Link ', # using the link column from Sora
                                        right_on='link') # and the link column from Ryan
                    .drop(['link', 'linkWithDate','numWords'], axis = 1) # get rid of columns I don't need 


negro_complete_df.head()
```

```python
# let's take a sample of the English texts and see how they're being read by the OCR:


negro_complete_df.sample(5)['paragraphs']
```

```python
# let's find the areas where "negro" is found: 

def find_negro(string):
    arr = string.split(sep = ' ') # split along the white space as an array
    neg_arg = np.argwhere('negro' in arr)
    return neg_arg
```
